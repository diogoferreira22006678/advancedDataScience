# Ames Housing Dataset – README Validation

This notebook validates the feature descriptions and characteristics documented in the README by exploring the actual dataset.

## 1. Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
sns.set_style('whitegrid')
```

## 2. Load and Inspect the Dataset

```python
# Load the dataset
df = pd.read_csv('../data/raw/AmesHousing.csv')

# Display basic information
print(f"Dataset shape: {df.shape}")
print(f"\nNumber of observations: {df.shape[0]}")
print(f"Number of features: {df.shape[1]}")
print(f"\nExpected: 81 explanatory variables + 1 target (SalePrice) = 82 total columns")
print(f"Note: Including Order and PID as part of the 81 explanatory variables")
```

## 3. Verify Column Names and Structure

```python
# List all columns
print(f"Total columns: {len(df.columns)}\n")
print("All columns:")
for i, col in enumerate(df.columns, 1):
    print(f"{i:2d}. {col}")
```

## 4. Inspect Identification Variables

```python
# Check identification columns
print("Identification Variables:")
print(f"\nOrder (first 10):")
print(df['Order'].head(10).values)
print(f"\nPID (first 10):")
print(df['PID'].head(10).values)
print(f"\nOrder dtype: {df['Order'].dtype}")
print(f"PID dtype: {df['PID'].dtype}")
print(f"\nOrder unique count: {df['Order'].nunique()}")
print(f"PID unique count: {df['PID'].nunique()}")
```

## 5. Validate Target Variable - SalePrice

```python
# Target variable analysis
print("Target Variable - SalePrice:")
print(f"Data type: {df['SalePrice'].dtype}")
print(f"\nDescriptive Statistics:")
print(df['SalePrice'].describe())
print(f"\nMissing values: {df['SalePrice'].isna().sum()}")
```

## 6. Sale Information Variables

```python
# Sale information
print("Sale Information Variables:")

print(f"\nMo Sold (Month):")
print(f"Type: {df['Mo Sold'].dtype}")
print(f"Range: {df['Mo Sold'].min()} - {df['Mo Sold'].max()}")
print(f"Unique values: {sorted(df['Mo Sold'].unique())}")

print(f"\nYr Sold (Year):")
print(f"Type: {df['Yr Sold'].dtype}")
print(f"Range: {df['Yr Sold'].min()} - {df['Yr Sold'].max()}")
print(f"Unique values: {sorted(df['Yr Sold'].unique())}")

print(f"\nSale Type (Nominal):")
print(f"Type: {df['Sale Type'].dtype}")
print(f"Unique values: {sorted(df['Sale Type'].unique())}")
print(f"\nValue counts:")
print(df['Sale Type'].value_counts())

print(f"\nSale Condition (Nominal):")
print(f"Type: {df['Sale Condition'].dtype}")
print(f"Unique values: {sorted(df['Sale Condition'].unique())}")
```

## 7. Location and Zoning Variables

```python
# Location and Zoning
print("Location and Zoning Variables:")

print(f"\nMS SubClass (Type of dwelling):")
print(f"Type: {df['MS SubClass'].dtype}")
print(f"Unique count: {df['MS SubClass'].nunique()}")
print(f"Sample values: {sorted(df['MS SubClass'].unique())[:10]}")

print(f"\nMS Zoning (General zoning):")
print(f"Type: {df['MS Zoning'].dtype}")
print(f"Unique values: {sorted(df['MS Zoning'].unique())}")

print(f"\nNeighborhood:")
print(f"Type: {df['Neighborhood'].dtype}")
print(f"Unique count: {df['Neighborhood'].nunique()}")
print(f"Sample neighborhoods: {sorted(df['Neighborhood'].unique())[:5]}")

print(f"\nCondition 1 & 2:")
print(f"Condition 1 unique: {sorted(df['Condition 1'].unique())}")
print(f"Condition 2 unique: {sorted(df['Condition 2'].unique())}")
```

## 8. Lot Characteristics Variables

```python
# Lot characteristics
lot_vars = ['Lot Frontage', 'Lot Area', 'Lot Shape', 'Land Contour', 'Lot Config', 'Land Slope']
print("Lot Characteristics Variables:")
print(df[lot_vars].describe())

print(f"\nLot Shape (Ordinal):")
print(f"Unique values: {sorted(df['Lot Shape'].unique())}")

print(f"\nLand Contour (Nominal):")
print(f"Unique values: {sorted(df['Land Contour'].unique())}")

print(f"\nLot Config (Nominal):")
print(f"Unique values: {sorted(df['Lot Config'].unique())}")

print(f"\nLand Slope (Ordinal):")
print(f"Unique values: {sorted(df['Land Slope'].unique())}")
```

## 9. House Style and Construction Variables

```python
# House style and construction
construction_vars = ['Overall Qual', 'Overall Cond', 'Year Built', 'Year Remod/Add']
print("House Style and Construction Variables:")
print(df[construction_vars].describe())

print(f"\nBldg Type (Nominal):")
print(f"Unique values: {sorted(df['Bldg Type'].unique())}")

print(f"\nHouse Style (Nominal):")
print(f"Unique values: {sorted(df['House Style'].unique())}")

print(f"\nRoof Style (Nominal):")
print(f"Unique values: {sorted(df['Roof Style'].unique())}")

print(f"\nRoof Matl (Nominal):")
print(f"Unique values: {sorted(df['Roof Matl'].unique())}")
```

## 10. Basement Variables

```python
# Basement variables
basement_vars = ['BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF', 'Total Bsmt SF', 'Bsmt Full Bath', 'Bsmt Half Bath']
print("Basement Variables (Continuous & Discrete):")
print(df[basement_vars].describe())

print(f"\nBsmt Qual (Ordinal):")
print(f"Unique values: {sorted(df['Bsmt Qual'].dropna().unique())}")

print(f"\nBsmt Cond (Ordinal):")
print(f"Unique values: {sorted(df['Bsmt Cond'].dropna().unique())}")

print(f"\nBsmt Exposure (Ordinal):")
print(f"Unique values: {sorted(df['Bsmt Exposure'].dropna().unique())}")

print(f"\nBsmtFin Type 1 (Ordinal):")
print(f"Unique values: {sorted(df['BsmtFin Type 1'].dropna().unique())}")
```

## 11. Heating and Electrical Variables

```python
# Heating and electrical
print("Heating and Electrical Variables:")

print(f"\nHeating (Nominal):")
print(f"Unique values: {sorted(df['Heating'].dropna().unique())}")

print(f"\nHeating QC (Ordinal):")
print(f"Unique values: {sorted(df['Heating QC'].dropna().unique())}")

print(f"\nCentral Air (Nominal):")
print(f"Unique values: {sorted(df['Central Air'].dropna().unique())}")

print(f"\nElectrical (Nominal):")
print(f"Unique values: {sorted(df['Electrical'].dropna().unique())}")
```

## 12. Living Area Variables

```python
# Living area
living_vars = ['1st Flr SF', '2nd Flr SF', 'Low Qual Fin SF', 'Gr Liv Area', 
               'Full Bath', 'Half Bath', 'Bedroom AbvGr', 'Kitchen AbvGr', 'TotRms AbvGrd']
print("Living Area Variables:")
print(df[living_vars].describe())

print(f"\nKitchen Qual (Ordinal):")
print(f"Unique values: {sorted(df['Kitchen Qual'].unique())}")

print(f"\nFunctional (Ordinal):")
print(f"Unique values: {sorted(df['Functional'].unique())}")
```

## 13. Fireplace Variables

```python
# Fireplace
print("Fireplace Variables:")
print(f"\nFireplaces (Discrete numerical):")
print(f"Type: {df['Fireplaces'].dtype}")
print(f"Range: {df['Fireplaces'].min()} - {df['Fireplaces'].max()}")
print(f"Value counts:")
print(df['Fireplaces'].value_counts().sort_index())

print(f"\nFireplace Qu (Ordinal):")
print(f"Unique values: {sorted(df['Fireplace Qu'].dropna().unique())}")
print(f"Value counts:")
print(df['Fireplace Qu'].value_counts())
```

## 14. Garage Variables

```python
# Garage
garage_vars = ['Garage Yr Blt', 'Garage Cars', 'Garage Area']
print("Garage Variables:")
print(df[garage_vars].describe())

print(f"\nGarage Type (Nominal):")
print(f"Unique values: {sorted(df['Garage Type'].dropna().unique())}")

print(f"\nGarage Finish (Ordinal):")
print(f"Unique values: {sorted(df['Garage Finish'].dropna().unique())}")

print(f"\nGarage Qual (Ordinal):")
print(f"Unique values: {sorted(df['Garage Qual'].dropna().unique())}")

print(f"\nGarage Cond (Ordinal):")
print(f"Unique values: {sorted(df['Garage Cond'].dropna().unique())}")
```

## 15. Porch and Outdoor Variables

```python
# Porch and outdoor
porch_vars = ['Paved Drive', 'Wood Deck SF', 'Open Porch SF', 'Enclosed Porch', 
              '3Ssn Porch', 'Screen Porch', 'Pool Area', 'Misc Val']
print("Porch and Outdoor Variables:")
print(df[porch_vars].describe())

print(f"\nPool QC (Ordinal):")
print(f"Unique values: {sorted(df['Pool QC'].dropna().unique())}")

print(f"\nFence (Ordinal):")
print(f"Unique values: {sorted(df['Fence'].dropna().unique())}")

print(f"\nMisc Feature (Nominal):")
print(f"Unique values: {sorted(df['Misc Feature'].dropna().unique())}")
```

## 16. Data Type Summary

```python
# Summary of data types
print("Data Type Summary:")
print(f"\nContinuous variables (float64/int64 with range):")
continuous = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
print(f"Count: {len(continuous)}")
print(f"Examples: {continuous[:10]}")

print(f"\nCategorical variables (object):")
categorical = df.select_dtypes(include=['object']).columns.tolist()
print(f"Count: {len(categorical)}")
print(f"Variables: {categorical}")

print(f"\nMissing values by column:")
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({'Count': missing, 'Percentage': missing_pct})
print(missing_df[missing_df['Count'] > 0].sort_values('Count', ascending=False))
```
